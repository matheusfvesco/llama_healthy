```python
import pandas as pd
```

```python
srag24 = pd.read_csv('data/INFLUD24-04-11-2024.csv', low_memory=False, sep=";")
```

```python
main_columns = [
    
    # Hospital
    'CLASSI_FIN',
    'CRITERIO',
    'EVOLUCAO',
    'HOSPITAL',
    'UTI',
    'SUPORT_VEN',
    
    # Demographics
    'DT_NOTIFIC',
    'DT_SIN_PRI',
    'DT_NASC',
    'SG_UF_NOT',
    'CS_ZONA',
    'CS_SEXO',
    'NU_IDADE_N',
    'CS_GESTANT',
    
    # Symptoms
    'FEBRE',
    'GARGANTA',
    'DISPNEIA',
    'DESC_RESP',
    'SATURACAO',
    'DIARREIA',
    'VOMITO',

    # Comorbidities
    'CARDIOPATI',
    'HEMATOLOGI',
    'SIND_DOWN',
    'HEPATICA',
    'ASMA',
    'DIABETES',
    'NEUROLOGIC',
    'PNEUMOPATI',
    'IMUNODEPRE',
    'RENAL',
    'OBESIDADE'
]


srag24_filt = srag24[main_columns]
```

```python
"""
    CRITERIO
    1. laboratorial
    2. clinico_epidemiologico
    3. clinico
    4. clinico_imagem
"""

def criterio(data):
    match data:
        case 1:
            return 'laboratorial'
        case 2:
            return 'clinico_epidemiologico'
        case 3:
            return 'clinico'
        case 4:
            return 'clinico_imagem'
        case _:
            return 'desconhecido'

# Aplicando a função para criar a coluna 'Doenca' no DataFrame
srag24_filt['criterio'] = srag24_filt['CRITERIO'].map(criterio)
```

```python
"""
    CLASSI_FIN
    1-influenza
    2-outro_virus_respiratório
    3-outro_agente_etiologico
    4-nao_especificado
    5-covid-19
"""
"""
    EVOLUCAO
    1-cura
    2-obito
    3-obito_por_outras_causas
    9-ignorado
"""
"""
    HOSPITAL
    1-hospitalizado
    2-nao_hospitalizado
    9-ignorado
"""
"""
    UTI
    1-sim
    2-nao
    9-ignorado
"""
"""
    SUPORT_VEN
    1-invasivo
    2-nao_invasivo
    3-sem_suporte_ventilatorio
    9-ignorado
"""
```

```python
# Função para mapear CLASSI_FIN
def classi_fin(data):
    match data:
        case 1:
            return 'influenza'
        case 2:
            return 'outro_virus_respiratorio'
        case 3:
            return 'outro_agente_etiologico'
        case 4:
            return 'nao_especificado'
        case 5:
            return 'covid-19'
        case _:
            return 'desconhecido'

# Função para mapear EVOLUCAO
def evolucao(data):
    match data:
        case 1:
            return 'cura'
        case 2:
            return 'obito'
        case 3:
            return 'obito_por_outras_causas'
        case 9:
            return 'ignorado'
        case _:
            return 'desconhecido'

# Função para mapear HOSPITAL
def hospital(data):
    match data:
        case 1:
            return 'hospitalizado'
        case 2:
            return 'nao_hospitalizado'
        case 9:
            return 'ignorado'
        case _:
            return 'desconhecido'

# Função para mapear UTI
def uti(data):
    match data:
        case 1:
            return 'sim'
        case 2:
            return 'nao'
        case 9:
            return 'ignorado'
        case _:
            return 'desconhecido'

# Função para mapear SUPORT_VEN
def suport_ven(data):
    match data:
        case 1:
            return 'invasivo'
        case 2:
            return 'nao_invasivo'
        case 3:
            return 'sem_suporte_ventilatorio'
        case 9:
            return 'ignorado'
        case _:
            return 'desconhecido'

# Aplicando as funções para criar as colunas mapeadas no DataFrame
srag24_filt['classi_fin'] = srag24_filt['CLASSI_FIN'].map(classi_fin)
srag24_filt['evolucao'] = srag24_filt['EVOLUCAO'].map(evolucao)
srag24_filt['hospital'] = srag24_filt['HOSPITAL'].map(hospital)
srag24_filt['uti'] = srag24_filt['UTI'].map(uti)
srag24_filt['suport_ven'] = srag24_filt['SUPORT_VEN'].map(suport_ven)
```

```python
sex_columns_dict = {
    "F" : "feminino",
    "M" : "masculino",
    "I" : "ignorado"
}

srag24_filt['CS_SEXO'] = srag24_filt['CS_SEXO'].map(sex_columns_dict)

srag_filt = srag24_filt
```

```python
# Symptoms and comorbidities and vaccination
sympt_comor_dict = dict.fromkeys([1], 1) 
sympt_comor_dict.update(dict.fromkeys([2,9], 0)) 

srag24_filt.replace({"FEBRE": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"GARGANTA": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"DISPNEIA": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"DESC_RESP": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"SATURACAO": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"DIARREIA": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"VOMITO": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"CARDIOPATI": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"HEMATOLOGI": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"SIND_DOWN": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"HEPATICA": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"ASMA": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"DIABETES": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"NEUROLOGIC": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"PNEUMOPATI": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"IMUNODEPRE": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"RENAL": sympt_comor_dict}, inplace=True)
srag24_filt.replace({"OBESIDADE": sympt_comor_dict}, inplace=True)

srag24_filt[['DT_NOTIFIC','DT_SIN_PRI', 'DT_NASC']] = srag24_filt[['DT_NOTIFIC','DT_SIN_PRI', 'DT_NASC']].apply(pd.to_datetime, format="%d/%m/%Y")

srag24_filt = srag24_filt.reset_index()
```
